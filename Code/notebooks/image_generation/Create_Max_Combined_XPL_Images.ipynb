```python
import os
import math
import cv2
import numpy as np
import pandas as pd
from PIL import Image, ImageOps, ImageDraw
from matplotlib import pyplot as plt
from copy import copy
from pathlib import Path
from shutil import copyfile
%matplotlib inline
```

```python
input_folder = f"./img_data_output"
```

```python
output_folder_name = "img_data_max"
```

```python
output_folder_path = f"./{output_folder_name}"
```

```python
# create output folder if not exist
Path(output_folder_path).mkdir(parents=True, exist_ok=True)
```

```python
rotation_ids = ['001', '009', '017', '025', '033', '041', '049', '057', '065']
section_ids = ["r1", "r2"]
```

```python
def show(img):
    #Show the image with matplotlib
    plt.imshow(img)
    plt.show()
```

```python
for subdir, dirs, files in os.walk(input_folder):
    for directory in dirs:
        for section_id in section_ids:
            # img.shape: (326, 326, 3)
            image_stacks = np.zeros(shape=(9, 326, 326, 3), dtype=np.uint8) # this consideres 9 rotational images
            for i in range(image_stacks.shape[0]):
                img_path = f"./img_data_output\{directory}\\{section_id}\{directory}_{section_id}_xpl{rotation_ids[i]}.jpg"
                #print(img_path)
                img_src = cv2.imread(img_path)
                img = cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB)
                image_stacks[i] = np.array(img)
            image_max = np.max(image_stacks, axis=0).astype(np.uint8)
            stacked_image = Image.fromarray(image_max)
            max_file_name = f"{directory}_{section_id}_xpl_max.jpg"
            print(max_file_name)
            #show(stacked_image)
            stacked_image.save(os.path.join(output_folder_path, max_file_name), "JPEG", quality=100)
```
