```python
import os
import math
import cv2
import shutil
import numpy as np
import pandas as pd
from PIL import Image, ImageOps, ImageDraw
from matplotlib import pyplot as plt
from copy import copy
from pathlib import Path
from shutil import copyfile
%matplotlib inline
```

```python
input_folder = os.path.join("Virtual_Microscope_Apollo_Images", "altered_img_data", "img_data_max_pixel_values_circular")
```

```python
output_folder = os.path.join("Virtual_Microscope_Apollo_Images", "altered_img_data", "img_data_max_pixel_values_circular_sharpened")
```

```python
def unsharp_mask(image, kernel_size=(5, 5), sigma=1.0, amount=1.0, threshold=0):
    """Return a sharpened version of the image, using an unsharp mask."""
    blurred = cv2.GaussianBlur(image, kernel_size, sigma)
    sharpened = float(amount + 1) * image - float(amount) * blurred
    sharpened = np.maximum(sharpened, np.zeros(sharpened.shape))
    sharpened = np.minimum(sharpened, 255 * np.ones(sharpened.shape))
    sharpened = sharpened.round().astype(np.uint8)
    if threshold > 0:
        low_contrast_mask = np.absolute(image - blurred) < threshold
        np.copyto(sharpened, image, where=low_contrast_mask)
    return sharpened
```

```python
try:
    Path(output_folder).mkdir(parents=True, exist_ok=False)
except:
    shutil.rmtree(output_folder)
    Path(output_folder).mkdir(parents=True, exist_ok=False)

for subdir, dirs, files in os.walk(input_folder):
    for file in files:
        img_src = cv2.imread(os.path.join(input_folder, file))
        #img = cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB)
        sharpened_image = unsharp_mask(img_src)
        #print(os.path.join(input_folder, file))
        cv2.imwrite(os.path.join(output_folder, file), sharpened_image)
        print(f"File {file} processed.")
        
```
