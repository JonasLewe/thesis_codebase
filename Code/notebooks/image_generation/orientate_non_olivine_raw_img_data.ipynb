```python
import os
import imutils
import shutil
import math
import cv2
import numpy as np
import pandas as pd
from PIL import Image, ImageOps, ImageDraw
from matplotlib import pyplot as plt
from pathlib import Path
from os import listdir
from shutil import copyfile
```

```python
input_folder = os.path.join("Virtual_Microscope_Other_Images", "raw_img_data")
```

```python
output_folder_path = os.path.join("Virtual_Microscope_Other_Images", "altered_img_data", "raw_img_data_rectified")
```

```python
def show(img):
    #Show the image with matplotlib
    plt.imshow(img)
    plt.show()
```

```python
def convert_to_pil(cv2_img):
    return Image.fromarray(cv2_img)
```

```python
def rotate_img_imutils(cv2_img, degrees):
    return convert_to_pil(imutils.rotate_bound(cv2_img, -degrees))
```

```python
def rotate_and_mask(cv2_image, degrees): # cv2 image as input
    circle_radius = cv2_image.shape[0]//2 - 5
    
    img = rotate_img_imutils(cv2_image, degrees) # outputs pil image
    
    def get_final_img(img):
        img_radius = img.size[0]//2

        upper_left = img_radius - circle_radius
        upper_left_tuple = (upper_left*3, upper_left*3)
        lower_right = img_radius + circle_radius
        lower_right_tuple = (lower_right*3, lower_right*3)

        # apply mask to img
        bigsize = (img.size[0] * 3, img.size[1] * 3)
        mask = Image.new('L', bigsize, 0)
        draw = ImageDraw.Draw(mask) 
        draw.ellipse(upper_left_tuple + lower_right_tuple, fill=255)
        mask = mask.resize(img.size, Image.ANTIALIAS)
        img.putalpha(mask)

        # convert alpha channel to white
        background = Image.new("RGB", img.size, (255, 255, 255))
        background.paste(img, mask = img.split()[3])

        # crop img to original size
        crop_area = (upper_left, upper_left, lower_right, lower_right)
        return background.crop(crop_area)
    
    img = get_final_img(img)
        
    return img # returns cropped pil images
```

```python
#try:
#    Path(output_folder_path).mkdir(parents=True, exist_ok=False)
#except:
#    shutil.rmtree(output_folder_path)
#    Path(output_folder_path).mkdir(parents=True, exist_ok=False)
for subdir, dirs, files in os.walk(input_folder):
    for file in files:
        specimen_number = file.split('_')[0]
        section_number = file.split('_')[1]
        # check if folders already exist
        Path(f"{output_folder_path}/{specimen_number}").mkdir(parents=True, exist_ok=True)
        Path(f"{output_folder_path}/{specimen_number}/{section_number}").mkdir(parents=True, exist_ok=True)
        if "_ppl0" in file:
            identifier = "ppl0"
            print("ppl")
        elif "_xpl0" in file:
            identifier = "xpl0"
            print("xpl")
        rotation_multiple = file.split('_')[-1].split('.')[0].replace(identifier, '')
        if rotation_multiple[0] == '0': # if single digit number
            rotation_multiple = rotation_multiple.replace('0', '')
        rotation_multiple = int(rotation_multiple)
        rotation_degrees = (rotation_multiple - 1) * 5
        img_src = cv2.imread((os.path.join(subdir, file)))
        img_src = cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB) # important so images have the right colors!
        img = rotate_and_mask(img_src, rotation_degrees)
        img.save(os.path.join(output_folder_path, specimen_number, section_number, file), "JPEG", quality=100)
        #show(img)
        print(f"File {file} processed.")
```
